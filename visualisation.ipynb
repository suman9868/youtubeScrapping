```python
data['Title'] = data['Title'].map(lambda com: com.split(" "))
data['Description'] = data['Description'].map(lambda com: com.split(" "))
# print(tt)
print(data['Title'].head(5))
print(data['Description'].head(5))
```

# wordcloud for each Category

```python
# fig = plt.figure(1)
columns = ['travel', 'science', 'food', 'manufacturing', 'history','art']
for col in columns:
    wordcloud = WordCloud(
                          background_color='white',
                          max_words=150,
                          max_font_size=80, 
                          random_state=10
                         ).generate(str(data[data['Category'] == col][['Title','Description']]))
    
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title('\n' + "Class: "+ col, fontsize=18, fontweight='bold')
    plt.axis('off')
    plt.show()
    #fig = plt.figure(1)
    #plt.savefig(col+".png", dpi=900)
    #break
```
